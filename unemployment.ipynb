#Compare Employed Vs. Unemployed

```
import plotly.graph_objects as go

fig = go.Figure(data=[
                      go.Bar(name='Employed',x=emp_20.State, y=emp_20.Employed ),
                      go.Bar(name='Unemployed',x=emp_20.State, y = emp_20.Unemployed )
])

#update the bar mode
fig.update_layout(barmode = 'group')
fig.show()
```

#Compare Employed Vs. Unemployed Vs. Employable

```
import plotly.graph_objects as go

fig = go.Figure(data=[
                      go.Bar(name='Employed',x=emp_20.State, y=emp_20.Employed ),
                      go.Bar(name='Unemployed',x=emp_20.State, y = emp_20.Unemployed ),
                      go.Bar(name='Employable_Population',x=emp_20.State, y = emp_20.Employable_Population )
])

#update the bar mode
fig.update_layout(barmode = 'group')
fig.show()
```

#stacked bar chart for Employed/Unemployed/Employable

```
import plotly.graph_objects as go

fig = go.Figure(data=[
                      go.Bar(name='Employed',x=emp_20.State, y=emp_20.Employed ),
                      go.Bar(name='Unemployed',x=emp_20.State, y = emp_20.Unemployed ),
                      go.Bar(name='Employable_Population',x=emp_20.State, y = emp_20.Employable_Population ),
                      
])

#update the bar mode
fig.update_layout(title_text='In 2020 Employed vs. Unemployed vs. Employable',barmode = 'stack')
fig.show()
```

# Density Heat map with histogram

```
fig = px.density_heatmap(emp_20, y = 'Employable_Population', x = 'Employed', 
                         marginal_x="histogram",
                         marginal_y="histogram",
                         width = 1000, height = 600, title= 'Heat Map/Histogram')
fig.show()
```
